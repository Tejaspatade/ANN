```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

def linear(x):
    return x

# Define input range
x = np.linspace(-5, 5, 100)

# Plot activation functions



```

```python
x = np.linspace(-10,10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Linear Activation Function')
plt.show()
```

```python
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
plt.plot(x, relu(x), label='ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python

plt.plot(x, tanh(x), label='Tanh')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Tanh')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python

plt.plot(x, softmax(x), label='Softmax')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Softmax')

# Add legend
plt.legend()

# Show plot
plt.show()
```
